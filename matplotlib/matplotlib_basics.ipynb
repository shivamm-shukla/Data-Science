# **2D Line Plots**

*   Bivariate Analysis
*   categorical vs numerical and numerical vs numerical
*   Used in time series data



```python
# plotting a simple function

price= [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'], batsman['V Kohli'], color='red')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle= 'dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted', linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle= 'dashed', linewidth=3, marker='o', markersize=5)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted', linewidth=2, marker='D', markersize=3)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend , location
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle= 'dashed', linewidth=3, marker='o', markersize=5, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted', linewidth=2, marker='D', markersize=3, label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# to show the label you need to call legend
# legend , location
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle= 'dashed', linewidth=3, marker='o', markersize=5, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted', linewidth=2, marker='D', markersize=3, label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
# you can pass loc for location of label
# legend , location
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle= 'dashed', linewidth=3, marker='o', markersize=5, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted', linewidth=2, marker='D', markersize=3, label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)

# in this graph, due to a very large value (here 4500000) variations for some points are showing
# so we use limiting axes to limit the range to see the variation in that part

plt.ylim(0,75000)
```

# **Scatter Plots**

*   Bivariate Analysis
*   Numerical vs numerical
*   Used in finding correlation



```python
# plt.scatter simple function

x = np.linspace(-10, 10, 50)
y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('/content/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'], color = 'red', marker = 'o')
plt.title('Avg and SR analysis of Top 50 batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# size
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size'])
```

```python
# we are scaling the size to visualize
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*20)
```

```python
# scatterplot using plt.plot

plt.plot(tips['total_bill'], tips['tip'])
```

```python
# This is faster than ploting scatter plot using plt.scatter
plt.plot(tips['total_bill'], tips['tip'],'o')
```

# **Bar Chart**

*  Bivariate Analysis
*  Numerical vs Categorical
*  Used in Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red', 'blue', 'green','yellow','pink']
plt.bar(colors,children, color='blue')
```

```python
# Horizontal bar chart

plt.barh(colors, children, color='red')
```

```python
# color and label

df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
# Multiple bar charts
plt.bar(df['batsman'], df['2015'])
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])

# isme kya hua ki phle 2015 wala print hua aur fir uske upar 2016 wala print ho gya hai
```

```python
plt.bar(df['batsman'], df['2015'], width=0.5)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman']) # to replace the label from existing to new one

plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# name overlaping issue
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='red')


plt.xticks(rotation='vertical')
```

```python
# Stacked Bar Chart

plt.bar(df['batsman'],df['2015'], label = '2015')
plt.bar(df['batsman'],df['2016'], bottom = df['2015'], label = '2016')
plt.bar(df['batsman'],df['2017'], bottom = df['2015'] + df['2016'], label = '2017')

plt.legend()
plt.show()
```

# **Pie Chart**

*  Univariable/Bivariate Analysis
*  Categorical vs numerical
*  used to find contribution on a standard scale

```python
# simple data
data = [23,45, 100, 20,49]
subjects = ['eng', 'science', 'maths', 'sst', 'hindi']
plt.pie(data, labels=subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
