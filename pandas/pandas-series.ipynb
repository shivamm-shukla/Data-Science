```python
import numpy as np
import pandas as pd
```

# **Series from lists**


```python
# String

country = ['India', 'Pakistan', 'USA', 'Nepal', 'Srilanka']
print(pd.Series(country))
```

```python
# Integers

runs = [12,23,35,55,56]
pd.Series(runs)
```

```python
# Custom Index

marks = [67,57,89,100]
subjects = ['maths','english', 'science', 'hindi']
pd.Series(marks, index=subjects)
```

```python
# Setting name

marks = pd.Series(marks,index=subjects,name="Shivam shukla's marks")
marks.name
```

# **Series from dict**

```python
marks = {
    'maths': 67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks, name= "Shivam Shukla's marks")
marks_series
```
